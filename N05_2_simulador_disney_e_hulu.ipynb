### Qual a mediana de duração dos filmes da Disney?

```python
df_disney_movies = df_disney[df_disney['type'] == 'Movie'].copy()
df_disney_movies.head()
```

```python
df_disney_movies['duration'] = (
    df_disney_movies['duration']
    .astype(str)
    .str.replace('min', '', regex=False)
)
df_disney_movies['duration'] = pd.to_numeric(df_disney_movies['duration'])
df_disney_movies['duration'].median()
```

### Quantos filmes na Disney foram lançados em 2013

```python
movies_2013 = df_disney_movies[df_disney_movies['release_year'] == 2013]
movies_2013.head().shape[0]
```
