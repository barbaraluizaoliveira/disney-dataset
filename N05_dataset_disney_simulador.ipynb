## Desvio padr√£o e m√©dia

```python

df_disney_movies = df_disney[df_disney['type'] == 'Movie'].copy()
```

```python
# tirando a m√©dia da dura√ß√£o dos filmes
df_disney_movies.head()
```

```python
# calculo de m√©dia com base em uma coluna, mas a duration aina n √© num√©rica
# vou tratar isso
# forma errada pois a duration coluna ainda n esta tratada ---------- df_disney_movies['duration'].mean()

# forma correta agr
df_disney_movies['duration'] = (
    df_disney_movies['duration']
    .astype(str)
    .str.replace('min', '', regex=False)
)
```

```python
df_disney_movies.head()
```

```python
df_disney_movies.dtypes
```

```python
#o type ainda continua obj mesmo que tenhamos tirado a palavra min
df_disney_movies['duration'] = df_disney_movies['duration'].astype(int)
```

```python
df_disney_movies['duration'].mean()
```
